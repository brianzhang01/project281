## Manhattan

```python
f = open('../data/manhattan_train.csv', 'rb')
freader = csv.reader(f)
data = list(freader)[1:]
lon, lat, counts = zip(*data)

lon_coords = [float(i) for i in lon]
lat_coords = [float(i) for i in lat]
counts = np.array([float(i) for i in counts])

loc = np.array(zip(lon_coords, lat_coords))
```

```python
poisson_distance = []
k_values = [1] + range(5,45,5) + range(45,100,5)
folds = 10

cross_valid = CrossValidation(folds=folds)
cross_valid.fit(loc, counts)

start = time.time()
for k in k_values:
    curr_scores = []
    for ind in xrange(folds):
        loc_train, loc_test, counts_train, counts_test = cross_valid.split_train_test(index=ind)
        neigh = KNeighbors(n_neighbors=k, weights='distance')
        neigh.fit(loc_train, counts_train)
    
        curr_scores.append(poisson_likelihood(neigh, loc_test, counts_test))

    poisson_distance.append(sum(curr_scores) / len(curr_scores))

poisson_uniform = []
for k in k_values:
    curr_scores = []
    for ind in xrange(folds):
        loc_train, loc_test, counts_train, counts_test = cross_valid.split_train_test(index=ind)
        
        neigh = KNeighbors(n_neighbors=k, weights='uniform')
        neigh.fit(loc_train, counts_train)
    
        curr_scores.append(poisson_likelihood(neigh, loc_test, counts_test))
    poisson_uniform.append(sum(curr_scores) / len(curr_scores))
```

```python
min_ll = float('inf')
min_k = None
min_weight = None

for k, value in zip(k_values, poisson_distance):
    if value < min_ll:
        min_ll = value
        min_k = k
        min_weight = 'distance'

for k, value in zip(k_values, poisson_uniform):
    if value < min_ll:
        min_ll = value
        min_k = k
        min_weight = 'uniform'
```

```python
print min_ll, min_k, min_weight
```

```python
# save knn results
f = open('knn_manhattan.csv', 'wb')
csvwriter = csv.writer(f, delimiter=',')

csvwriter.writerow(['k', 'poisson_uniform', 'poison_distance'])
for k, u, d in zip(k_values, poisson_uniform, poisson_distance):
    csvwriter.writerow([k,u,d])
f.close()
```
