```python
kriging = 13.01
average = 13.78
perfect = 2.117
knn_uniform = 10.355
knn_distance = 13.48158995

rcParams.update({'font.size': 12})
all_results = [perfect, average, kriging, knn_uniform, knn_distance]

loc = range(len(all_results))
plt.figure(figsize=(6,6))
plt.bar(loc, all_results, color='r')
ticks = plt.xticks([i + 0.5 for i in loc], ['perfect', 'average','kriging','KNN (unif)', 'KNN (dist)'], fontsize=14, rotation=30)
plt.xlabel('Model')
plt.ylabel('Negative Log Likelihood per point')
plt.title('Performance Between Models for Manhattan Data')
plt.show()
```
