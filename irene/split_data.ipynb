```python
import csv
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import poisson
import math
from random import randrange

import matplotlib
```

```python
# opens and reads all collisions data

f = open('../data/collisions_new.csv')
raw_data = csv.DictReader(f)
data = list(raw_data)
f.close()
print len(data)
data = [i for i in data if i['lat'] != '' and i['lon'] != '' and i['collisions'] != '']
print len(data)

# Just Manhattan
collisions_count = {}
for incident in data:
#     if int(incident['borocode']) == 1:
    lon = incident['lon']
    lat = incident['lat']
    boro = incident['borocode']
    c_count = incident['collisions']
    count = collisions_count.get((lon, lat, boro), 0)
    collisions_count[(lon, lat, boro)] = count + int(c_count)
```

```python
# create doc of all boros
g = open('../data/collisions_count_all.csv', 'wb')
gwriter = csv.writer(g, delimiter=',')

gwriter.writerow(['lon', 'lat', 'boro', 'count'])

for lon, lat, boro in collisions_count.keys():
    gwriter.writerow([lon, lat, boro, collisions_count[(lon, lat, boro)]])
g.close()
```

```python
# parse into boroughs, specifically Manhattan (1) and Brooklyn (3)

f = open('../data/manhattan_all.csv', 'wb')
fwriter = csv.writer(f)
fwriter.writerow(['lon', 'lat', 'count'])
for lon, lat, boro in collisions_count.keys():
    if boro == '1':
        fwriter.writerow([lon, lat, collisions_count[(lon, lat, boro)]])
f.close()

f = open('../data/brooklyn_all.csv', 'wb')
fwriter = csv.writer(f)
fwriter.writerow(['lon', 'lat', 'count'])
for lon, lat, boro in collisions_count.keys():
    if boro == '3':
        fwriter.writerow([lon, lat, collisions_count[(lon, lat, boro)]])
f.close()
```

```python
def shuffle(x):
    for i in xrange(len(x)-1, 0, -1):
        j = randrange(i + 1)
        x[i], x[j] = x[j], x[i]
    return x
```

```python
##### DO NOT RUN AGAIN WITHOUT MEANING TO #####
# create test and train data out of a given file and percentage
p = 0.25

g = open('../data/manhattan_all.csv', 'rb')
greader = csv.reader(g)
manhattan_data = list(greader)[1:]
g.close()

lon, lat, count = zip(*manhattan_data)
loc = zip(lon, lat)

n = len(loc)

shuffled_indicies = shuffle(range(n))
mid_point = int(p * n)

f = open('../data/manhattan_test.csv', 'wb')
fwriter = csv.writer(f)
fwriter.writerow(['lon', 'lat', 'count'])
for i in shuffled_indicies[:mid_point]:
    fwriter.writerow([loc[i][0], loc[i][1], count[i]])
f.close()

f = open('../data/manhattan_train.csv', 'wb')
fwriter = csv.writer(f)
fwriter.writerow(['lon', 'lat', 'count'])
for i in shuffled_indicies[mid_point:]:
    fwriter.writerow([loc[i][0], loc[i][1], count[i]])
f.close()
```

```python
# plot manhattan test and train data
FACTOR = 0.7581

f = open('../data/manhattan_test.csv', 'rb')
freader = csv.reader(f)
data = list(freader)[1:]
lon, lat, count = zip(*data)

lon_coords = [float(i) for i in lon]
lat_coords = [float(i) for i in lat]
count = [float(i) for i in count]
colors = [min(i*1.0/50, 1) for i in count]

fig = plt.figure(figsize=(12,20/FACTOR), dpi=100)
plt.scatter(lon_coords, lat_coords, c=colors)
#plt.gray()
plt.xlabel('longitude')
plt.ylabel('latitude')
# plt.xlim([-74.02,-73.9])
# plt.ylim([40.7, 40.9])
plt.grid(True)
plt.title('Preliminary Manhattan Test Data')
matplotlib.rcParams.update({'font.size': 14})
plt.show()
```
