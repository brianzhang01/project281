In our earlier model, the accuracy is determined by whether or not the prediction was correct. Instead, let's see if we can model using a Poisson distribution to plot log-likelihood.

```python
from scipy.stats import poisson

# rv = poisson(5.)
# print rv.pmf(8)
def poisson_score(kneigh, features, labels):
    predictions = kneigh.predict(features) 
    results = [-math.log(poisson(p).pmf(l)) for (p, l) in zip(predictions, labels)]
    return sum(results)
```

```python
poisson_scores = []
k_values = range(5,40,5)

for k in k_values:
    curr_scores = []
    for r in random_states:
        loc_train, loc_test, counts_train, counts_test = train_test_split(loc, counts, test_size=0.2, random_state=r)
        
        neigh = KNeighborsClassifier(n_neighbors=k, weights='distance')
        neigh.fit(loc_train, counts_train)
    
        curr_scores.append(poisson_score(neigh, loc_test, counts_test))
    poisson_scores.append(sum(curr_scores) / len(curr_scores))

plt.plot(k_values, poisson_scores)
```
