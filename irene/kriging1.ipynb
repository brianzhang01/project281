#### Implementing basic kriging in 1D for practice

Steps to do:

Define kernel functions (Dougal used 4)

Partition the interval (0, 1)

Use the Cholesky decomposition to make draws from a multivariate normal (can also use numpy.random.multivariate_normal)

```python
import math
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Let's vectorize
def matern_1(x,L):
    return np.exp(-x/L)
def matern_3(x,L):
    return (1+math.sqrt(3)*x/L)*np.exp(-math.sqrt(3)*x/L)
def matern_5(x,L):
    return (1+math.sqrt(5)*x/L+5*np.square(x)/(3*L*L))*np.exp(-math.sqrt(5)*x/L)
```

```python
x = np.arange(0, 1, 0.01)
plt.plot(x,matern_1(x,0.25),'r')
plt.plot(x,matern_3(x,0.16),'b')
plt.plot(x,matern_5(x,0.13),'g')
plt.show()
# why doesn't this match Figure 3.2 of Diggle / Ribeiro?
```

```python
x = np.arange(0, 1, 0.005)
N = len(x)
# slow way
#D = np.abs(np.tile(x,(N,1)) - np.tile(x,(N,1)).T)
# fast way
D = np.abs(x - x[:, None])
```

```python
K1 = matern_1(D,0.3)
y1 = np.random.multivariate_normal(np.zeros(N), K1)
K3 = matern_3(D,0.3)
y3 = np.random.multivariate_normal(np.zeros(N), K3)
K5 = matern_5(D,0.3)
y5 = np.random.multivariate_normal(np.zeros(N), K5)
plt.plot(x,y1,'r')
plt.plot(x,y3,'g')
plt.plot(x,y5,'b')
plt.show()
```

```python
y3a = np.random.multivariate_normal(np.zeros(N), matern_3(D,0.3))
y3b = np.random.multivariate_normal(np.zeros(N), 5*matern_3(D,0.3))
y3c = np.random.multivariate_normal(np.zeros(N), matern_3(D,1.5))
plt.plot(x,y3a,'r')
plt.plot(x,y3b,'g')
plt.plot(x,y3c,'b')
plt.show()
```

```python
samples = np.random.randint(N, size = 5)
xd = x[samples]
yd = y5[samples]
plt.scatter(xd,yd)
plt.show()
```

```python
def kernel(x):
    return matern_5(x,0.2)
```

```python
xp = np.arange(0, 1, 0.01)
Kdd = kernel(np.abs(xd - xd[:, None]))
Kpp = kernel(np.abs(xp - xp[:, None]))
Kpd = kernel(np.abs(xd - xp[:, None]))
```

```python
yp_mean = np.dot(np.dot(Kpd,np.linalg.inv(Kdd)),yd)
yp_sigma = np.diag(Kpp - np.dot(np.dot(Kpd,np.linalg.inv(Kdd)),Kpd.T))
```

#### Noise-free observations

```python
plt.plot(xp,yp_mean)
plt.plot(xp,yp_mean+yp_sigma,'r')
plt.plot(xp,yp_mean-yp_sigma,'r')
plt.scatter(xd,yd)
plt.show()
```

#### Noisy observations

```python
sigma_y = 0.5
sigma_f = 1
l = 0.2

def kernel(x):
    return matern_5(x,l)

xp = np.arange(0, 1, 0.01)
Kdd = sigma_f**2*kernel(np.abs(xd - xd[:, None]))+sigma_y**2*np.eye(np.size(xd))
Kpp = sigma_f**2*kernel(np.abs(xp - xp[:, None]))+sigma_y**2*np.eye(np.size(xp))
Kpd = sigma_f**2*kernel(np.abs(xd - xp[:, None]))
yp_mean = np.dot(np.dot(Kpd,np.linalg.inv(Kdd)),yd)
yp_sigma = np.diag(Kpp - np.dot(np.dot(Kpd,np.linalg.inv(Kdd)),Kpd.T))
plt.plot(xp,yp_mean)
plt.plot(xp,yp_mean+yp_sigma,'r')
plt.plot(xp,yp_mean-yp_sigma,'r')
plt.scatter(xd,yd)
plt.show()
```
