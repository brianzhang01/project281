```python
import csv
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
# from mpl_toolkits.basemap import Basemap
from scipy.stats import poisson
import math
from random import randrange
```

```python
f = open('../data/collisions_new.csv')
raw_data = csv.DictReader(f)
data = list(raw_data)
f.close()
print len(data)
data = [i for i in data if i['lat'] != '' and i['lon'] != '' and i['collisions'] != '']
print len(data)
```

```python
# Just Manhattan
collisions_count = {}
for incident in data:
    if int(incident['borocode']) == 1:
        lon = incident['lon']
        lat = incident['lat']
        c_count = incident['collisions']
        count = collisions_count.get((lon, lat), 0)
        collisions_count[(lon, lat)] = count + int(c_count)
```

```python
FACTOR = 0.7581
x_coords = []
y_coords = []
counts = []

for x,y in collisions_count.keys():
    x_coords.append(float(x))
    y_coords.append(float(y))
    counts.append(float(collisions_count[(x,y)]))

loc = np.transpose(np.array([x_coords,y_coords])) # original lats and lons, haven't applied scaling factor
counts = np.array(counts)
```

```python
print 'Total intersections: %d' % len(counts)
print 'Average collisions per interesection: %.2f' % (sum(counts) / float(len(counts)))

log_counts = np.log(counts)
plt.hist(log_counts, normed=False, bins=30)
plt.show()
```

```python
# Irene's cross-validation code
def shuffle(x):
    for i in xrange(len(x)-1, 0, -1):
        j = randrange(i + 1)
        x[i], x[j] = x[j], x[i]
    return x
        
def flatten_list(list_of_lists):
        return [item for sublist in list_of_lists for item in sublist]

def split_groups(lst, n):
    return [lst[i::n] for i in xrange(n)]

# k-fold cross validation
# usage: fit first, then call split_train_test(index=2) to get the different splittings of test/train data
class CrossValidation(object):
    def __init__(self, folds=5):
        self.folds = folds
    
    def flatten_list(self, list_of_lists):
        return [item for sublist in list_of_lists for item in sublist]
    
    def fit(self, locations, counts):
        self.locations = np.array(locations)
        self.counts = np.array(counts)
        self.cross_validation_data = {}
        self.n = len(locations)
        
        indicies = range(self.n)
        shuffled_indicies = shuffle(indicies)
        
        groups = np.array(split_groups(shuffled_indicies, self.folds))
        
        for fld in xrange(self.folds):
#            fold_data = {}
#            train_indicies = range(self.folds).remove(fld)
            fold_data = {}
            train_indicies = range(self.folds)
            train_indicies.remove(fld)

            test_indicies = fld
            train_set = self.flatten_list(groups[train_indicies])

#            train_set = self.flatten_list(groups[train_indicies][0])
            test_set = groups[test_indicies]
    
            fold_data['train'] = list(train_set)
            fold_data['test'] = test_set
            self.cross_validation_data[fld] = fold_data
        
    def split_train_test(self, index=0):
        if index >= self.folds:
            print 'Index too big'
            return
        else:
            fold_data = self.cross_validation_data[index]
            train = fold_data['train']
            test = fold_data['test']
            
            return self.locations[train], self.locations[test], self.counts[train], self.counts[test]
```

```python
# usage: create object, fit, predict
class Kriging(object):
    def __init__(self, kernel='matern_3', length_scale = 0.1, sigma_y = 1, sigma_f = 5):
        self.kernel = kernel
        self.length_scale = length_scale
        self.sigma_y = sigma_y
        self.sigma_f = sigma_f
        self.locations = None
        self.n = None
        self.yd = None
        self.FACTOR = 0.7581
        self.earth_radius = 3959 # in miles

    def fit(self, locations, counts):
        self.locations = locations
        self.n = len(locations)
        self.yd = counts

    def predict(self, new_sites):
        Kdd = self.build_kernel_matrix(self.locations, self.locations)
#        Kpp = self.build_kernel_matrix(new_sites, new_sites)
        Kpd = self.build_kernel_matrix(new_sites, self.locations)
        Kdd = self.sigma_f**2*Kdd + self.sigma_y**2*np.eye(np.size(self.locations)/2)
#        Kpp = self.sigma_f**2*Kpp + self.sigma_y**2*np.eye(np.size(new_sites)/2)
        Kpd = self.sigma_f**2*Kpd
        yp_mean = np.dot(np.dot(Kpd,np.linalg.inv(Kdd)),self.yd)
#        yp_sigma = np.diag(Kpp - np.dot(np.dot(Kpd,np.linalg.inv(Kdd)),Kpd.T))        
        return yp_mean

    def variances(self, new_sites):
        Kdd = self.build_kernel_matrix(self.locations, self.locations)
        Kpp = self.build_kernel_matrix(new_sites, new_sites)
        Kpd = self.build_kernel_matrix(new_sites, self.locations)
        Kdd = self.sigma_f**2*Kdd + self.sigma_y**2*np.eye(np.size(self.locations)/2)
        Kpp = self.sigma_f**2*Kpp + self.sigma_y**2*np.eye(np.size(new_sites)/2)
        Kpd = self.sigma_f**2*Kpd
#        yp_mean = np.dot(np.dot(Kpd,np.linalg.inv(Kdd)),self.yd)
        yp_vars = np.diag(Kpp - np.dot(np.dot(Kpd,np.linalg.inv(Kdd)),Kpd.T))        
        return yp_vars
        
    
    def build_kernel_matrix(self, locs1, locs2):
        return self.kernel_function(self.build_distance_matrix(locs1, locs2))

    def build_distance_matrix(self, locs1, locs2):
        lons1, lats1 = locs1[:,0], locs1[:,1]
        lons2, lats2 = locs2[:,0], locs2[:,1]
        dlon12 = lons2 - lons1[:, None]
        dlat12 = lats2 - lats1[:, None]
        D12 = np.sqrt(np.square(dlon12*self.FACTOR) + np.square(dlat12))
        D12 = D12/360*(2*math.pi)*self.earth_radius
        return D12

    def kernel_function(self, x):
        if self.kernel == 'matern_1':
            return self.matern_1(x,self.length_scale)
        elif self.kernel == 'matern_3':
            return self.matern_3(x,self.length_scale)
        else:
            return self.matern_5(x,self.length_scale)        
    
    def matern_1(self,x,L):
        return np.exp(-x/L)
    
    def matern_3(self,x,L):
        return (1+math.sqrt(3)*x/L)*np.exp(-math.sqrt(3)*x/L)
    
    def matern_5(self,x,L):
        return (1+math.sqrt(5)*x/L+5*np.square(x)/(3*L*L))*np.exp(-math.sqrt(5)*x/L)

# neigh = KNeighbors(n_neighbors=8, weights='uniform')
# neigh.fit(loc_train, counts_train)
# predictions = neigh.predict(loc_test)
```
