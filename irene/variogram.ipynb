```python
EARTH_RADIUS = 3959 # in miles

distances_x = np.subtract(x_coords, np.transpose(x_coords))
distances_y = np.subtract(y_coords, np.transpose(y_coords))
distances = (distances_x ** 2 + distances_y**2) ** (0.5)

distances_x = np.subtract(x_coords, np.transpose(x_coords))
distances_y = np.subtract(y_coords, np.transpose(y_coords))
distances = (distances_x ** 2 + distances_y**2) ** (0.5)
distances = distances /360*(2*math.pi)*EARTH_RADIUS

counts_var = np.subtract(counts, np.transpose(counts))
counts_var = 0.5*counts_var**2
```

```python
# from itertools import combinations

n = len(x_coords)

all_v = [counts_var[i,j] for i in xrange(n) for j in xrange(i)]
all_u = [distances[i,j] for i in xrange(n) for j in xrange(i)]
```

```python
small = 10000
plt.scatter(all_u[:small], all_v[:small])
plt.scatter(all_u[small:small+small], all_v[small:small+small])
plt.show()
# print len(all_u)
```

```python
def split_list(L, n):
    for i in range(0, L, n):
        yield i, i+n
```

```python
big = len(all_u)
bins = 30

partition_length = 100

values_weights = np.zeros(bins)
values_uniform = np.zeros(bins)

edges = np.histogram(all_u[:100], bins=bins)[1]

for start, stop in split_list(big, partition_length):
    values_weights += np.histogram(all_u[start:stop], bins=bins, weights=all_v[start:stop])[0]
    values_uniform += np.histogram(all_u[start:stop], bins=bins)[0]

bin_means = values_weights / values_uniform
bin_locs = [(edges[i] + edges[i+1]) / 2 for i in xrange(bins)]
```

```python
plt.scatter(bin_locs, bin_means)
plt.xlabel('pairwise distance (miles)')
plt.ylabel('variance of difference between collisions')
plt.title('Variogram of Manhattan data')
plt.show()

plt.bar(bin_locs, values_uniform, 0.48)
plt.xlabel('pairwise distance (miles)')
plt.ylabel('number of data points')
plt.title('Distribution of Manhattan pairwise distances')
plt.show()
```
