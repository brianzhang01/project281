```python
data = [i for i in data if i['lat'] != '' and i['lon'] != '' and i['collisions'] != '']
data[0]
```

```python
keys = ['lon', 'lat', 'collisions']
for key in keys:
    print data[0][key]
num_data = len(data)
print num_data
float(data[0]['lon'])
```

```python
x = np.zeros(num_data)
y = np.zeros(num_data)
collisions = np.zeros(num_data)
borocodes = np.zeros(num_data)
for i in range(num_data):
    x[i] = data[i]['lon']
    y[i] = data[i]['lat']
    collisions[i] = data[i]['collisions']
    borocodes[i] = data[i]['borocode']
```

```python
btocolor = {1: 'r', 2: 'b', 3: 'g', 4: 'c', 5: 'm'}
borocolors = [btocolor[i] for i in borocodes]
```

```python
plt.scatter(x, y)
plt.show()
```

```python
fig = plt.figure(figsize=(14,12), dpi=100)
plt.scatter(x, y)
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.xlim([-74.3, -73.6])
plt.ylim([40.4, 41.0])
plt.grid(True)
plt.title('Intersections')
plt.show()
```

```python
fig = plt.figure(figsize=(14,12), dpi=100)
plt.scatter(x, y, c=borocolors)
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.xlim([-74.3,-73.6])
plt.ylim([40.4, 41.0])
plt.grid(True)
plt.title('Intersections')
plt.show()
```
