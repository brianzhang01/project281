#### Kriging function

```python
# usage: create object, fit, predict
class Kriging(object):
    def __init__(self, kernel='matern_3', length_scale = 0.1, sigma_y = 1, sigma_f = 5):
        self.kernel = kernel
        self.length_scale = length_scale
        self.sigma_y = sigma_y
        self.sigma_f = sigma_f
        self.locations = None
        self.n = None
        self.yd = None
        self.FACTOR = 0.7581
        self.earth_radius = 3959 # in miles

    def fit(self, locations, counts):
        self.locations = locations
        self.n = len(locations)
        self.yd = counts

    def predict(self, new_sites):
        Kdd = self.build_kernel_matrix(self.locations, self.locations)
#        Kpp = self.build_kernel_matrix(new_sites, new_sites)
        Kpd = self.build_kernel_matrix(new_sites, self.locations)
        Kdd = self.sigma_f**2*Kdd + self.sigma_y**2*np.eye(np.size(self.locations)/2)
#        Kpp = self.sigma_f**2*Kpp + self.sigma_y**2*np.eye(np.size(new_sites)/2)
        Kpd = self.sigma_f**2*Kpd
        yp_mean = np.dot(np.dot(Kpd,np.linalg.inv(Kdd)),self.yd)
#        yp_sigma = np.diag(Kpp - np.dot(np.dot(Kpd,np.linalg.inv(Kdd)),Kpd.T))        
        return yp_mean

    def variances(self, new_sites):
        Kdd = self.build_kernel_matrix(self.locations, self.locations)
        Kpp = self.build_kernel_matrix(new_sites, new_sites)
        Kpd = self.build_kernel_matrix(new_sites, self.locations)
        Kdd = self.sigma_f**2*Kdd + self.sigma_y**2*np.eye(np.size(self.locations)/2)
        Kpp = self.sigma_f**2*Kpp + self.sigma_y**2*np.eye(np.size(new_sites)/2)
        Kpd = self.sigma_f**2*Kpd
#        yp_mean = np.dot(np.dot(Kpd,np.linalg.inv(Kdd)),self.yd)
        yp_vars = np.diag(Kpp - np.dot(np.dot(Kpd,np.linalg.inv(Kdd)),Kpd.T))        
        return yp_vars
        
    
    def build_kernel_matrix(self, locs1, locs2):
        return self.kernel_function(self.build_distance_matrix(locs1, locs2))

    def build_distance_matrix(self, locs1, locs2):
        lons1, lats1 = locs1[:,0], locs1[:,1]
        lons2, lats2 = locs2[:,0], locs2[:,1]
        dlon12 = lons2 - lons1[:, None]
        dlat12 = lats2 - lats1[:, None]
        D12 = np.sqrt(np.square(dlon12*self.FACTOR) + np.square(dlat12))
        D12 = D12/360*(2*math.pi)*self.earth_radius
        return D12

    def kernel_function(self, x):
        if self.kernel == 'matern_1':
            return self.matern_1(x,self.length_scale)
        elif self.kernel == 'matern_3':
            return self.matern_3(x,self.length_scale)
        else:
            return self.matern_5(x,self.length_scale)        
    
    def matern_1(self,x,L):
        return np.exp(-x/L)
    
    def matern_3(self,x,L):
        return (1+math.sqrt(3)*x/L)*np.exp(-math.sqrt(3)*x/L)
    
    def matern_5(self,x,L):
        return (1+math.sqrt(5)*x/L+5*np.square(x)/(3*L*L))*np.exp(-math.sqrt(5)*x/L)

# neigh = KNeighbors(n_neighbors=8, weights='uniform')
# neigh.fit(loc_train, counts_train)
# predictions = neigh.predict(loc_test)
```

#### Train-test performance for Manhattan

```python
from scipy.stats import poisson
def poisson_likelihood(p, c):
    try:
        NLL = -math.log(poisson(p).pmf(c))
    # handle k >>> 0 cases using Stirling's approximation
    except ValueError:
        print 'error'
        NLL = p - c*math.log(p) + c*math.log(c) - c + math.log(math.sqrt(2 * math.pi * c))
    return NLL
```

```python
# Running the Manhattan test, using parameters found from maximizing the marginal likelihood
kriger = Kriging(kernel='matern_3', length_scale=1.95, sigma_y = math.sqrt(1.34), sigma_f = math.sqrt(8.66))
kriger.fit(loc_train, np.log(counts_train))
log_pred_counts = kriger.predict(loc_test)
pred_counts = np.exp(log_pred_counts)
test_size = len(counts_test)
sum = 0.0
for i in range(test_size):
    sum = sum + poisson_likelihood(pred_counts[i], counts_test[i])
sum = sum / test_size
print sum
```

```python
nx = 100
ny = 200
m = Basemap(llcrnrlon=-74.03,llcrnrlat=40.69,urcrnrlon=-73.9,urcrnrlat=40.88,resolution='l',projection='merc',)
grid = m.makegrid(nx,ny)
loc_grid = np.column_stack((grid[0].ravel()[:,None],grid[1].ravel()[:,None]))
```

```python
kriger = Kriging(kernel='matern_3', length_scale=1.95, sigma_y = math.sqrt(1.34), sigma_f = math.sqrt(8.66))
kriger.fit(loc_train, np.log(counts_train))
grid_predictions = kriger.predict(loc_grid)
grid_predictions.resize((ny,nx))
```
