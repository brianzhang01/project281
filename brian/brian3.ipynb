#### Summary statistics

```python
borocounts = {}
collisions_count = {}
i = 0
for incident in data:
    lon = incident['lon']
    lat = incident['lat']
    c_count = incident['collisions']
    count = collisions_count.get((lon, lat), 0)
    collisions_count[(lon, lat)] = count + int(c_count)
borocounts[i] = np.copy(collisions_count.values())
for i in range(1,6):
    collisions_count = {}
    for incident in data:
        if int(incident['borocode']) == i:
            lon = incident['lon']
            lat = incident['lat']
            c_count = incident['collisions']
            count = collisions_count.get((lon, lat), 0)
            collisions_count[(lon, lat)] = count + int(c_count)
    borocounts[i] = np.copy(collisions_count.values())
```

```python
itoboro = {0: 'Overall', 1: 'Manhattan', 2: 'Brooklyn', 3: 'Queens', 4: 'The Bronx', 5: 'Staten Island'}
borototal = [len(borodata) for borodata in borocounts.values()]
boroavg = [(sum(borodata) / float(len(borodata))) for borodata in borocounts.values()]
for i in range(6):
    print 'Borough: %s' % itoboro[i]
    print 'Total intersections: %d' % borototal[i]
    print 'Average collisions per interesection: %.2f' % boroavg[i]
    print

# Overall total = 39040
# Sum of five boroughs = 39082

# TODO: check that different intersections is also different streets
```

#### Baselines

```python
N = len(counts)
mean = sum(counts)*1.0/len(counts)
print N
print mean
```

```python
rv = poisson(mean)
y_avg = -np.log(rv.pmf(counts))
avg_baseline = sum(y_avg)/N
print avg_baseline
```

```python
y_best = -np.log(poisson(counts).pmf(counts))
best_baseline = sum(y_best)/N
print best_baseline
```
