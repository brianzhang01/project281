```python
import csv
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from scipy.stats import poisson
import math
import matplotlib
```

```python
f = open('../data/brooklyn_train.csv', 'rb')
freader = csv.reader(f)
data = list(freader)[1:]
lon, lat, counts = zip(*data)

lon_coords = [float(i) for i in lon]
lat_coords = [float(i) for i in lat]
counts = np.array([float(i) for i in counts])

loc = np.array(zip(lon_coords, lat_coords))
```

```python
print 'Total intersections: %d' % len(counts)
print 'Average collisions per interesection: %.2f' % (sum(counts) / float(len(counts)))

plt.hist(counts, normed=True, bins=max(counts)-min(counts))
plt.axis([0, 30, 0, 0.25])
#plt.yscale('log')
```

```python
plt.scatter(lon_coords, lat_coords)
plt.show()
```

```python
colors = [min(i*1.0/25, 1) for i in counts]
# create new figure, axes instances.
fig=plt.figure(figsize=(15, 15))
ax=fig.add_axes([0.1,0.1,0.8,0.8])
m = Basemap(llcrnrlon=-74.05,llcrnrlat=40.55,urcrnrlon=-73.85,urcrnrlat=40.75,resolution='l',projection='merc',)
m.drawcounties(0.3)
m.drawparallels(np.arange(40.55,40.75,0.05),labels=[1,0,0,1])
m.drawmeridians(np.arange(-74.05,-73.85,0.1),labels=[1,0,0,1])
m.scatter(lon_coords, lat_coords, c=colors, latlon=True)
m.drawmapscale(lon=-73.90, lat=40.57, lon0=-73.94, lat0=40.66,
               barstyle='fancy', units='mi', length=4, fontsize=14, yoffset=(0.01*(m.ymax-m.ymin)))
cb = m.colorbar()
cb.set_ticks([.2, .4, .6, .8, 1])  # force there to be only 5 ticks
cb.set_ticklabels([5, 10, 15, 20, '>25'])  # put text labels on them
ax.set_title('Brooklyn collision count by intersection')
#plt.savefig('manhattan3.png', dpi=600)
plt.show()
```
